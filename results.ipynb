```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics
labels = [
    'Benign\nPrecision',
    'Benign\nRecall',
    'Benign\nF1',
    'Malignant\nPrecision',
    'Malignant\nRecall',
    'Malignant\nF1',
    'Accuracy'
]

values = [0.55, 0.29, 0.38, 0.64, 0.84, 0.73, 0.62]

# Different colors for each bar
colors = ['red', 'orange', 'gold', 'green', 'blue', 'purple', 'cyan']

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, values, color=colors)

# Add value labels on top of bars
for bar, value in zip(bars, values):
    plt.text(
        bar.get_x() + bar.get_width()/2,
        value + 0.02,
        f'{value:.2f}',
        ha='center',
        fontsize=10
    )

plt.ylim(0, 1.0)
plt.ylabel('Score')
plt.title('Classification Performance Metrics')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

### BREAST CANCER CLASSIFICATION USING ADVANCED DEEP LEARNING

#### STAGE-1: IMPORT LIBRARIES

```python
import os
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
from tensorflow.keras.applications import efficientnet_v2
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report, roc_curve
)
import itertools
import seaborn as sns
```

#### 2. DATA PREPROCESSING AND AUGMENTATION

```python
TRAIN_DIR  = r"C:\Users\srira\Downloads\Project\901\ultrasound breast classification dataset\train"
VALID_DIR  = r"C:\Users\srira\Downloads\Project\901\ultrasound breast classification dataset\val"
TEST_DIR   = r"C:\Users\srira\Downloads\Project\901\ultrasound breast classification dataset\test"
```

```python
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42
EPOCHS = 10
AUTOTUNE = tf.data.AUTOTUNE
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    TRAIN_DIR,
    labels='inferred',
    label_mode='binary',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=SEED
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    VALID_DIR,
    labels='inferred',
    label_mode='binary',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False,
    seed=SEED
)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    TEST_DIR,
    labels='inferred',
    label_mode='binary',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False,
    seed=SEED
)
```

```python
def show_sample_images(dataset, title, class_names=None):
    plt.figure(figsize=(10, 10))
    for images, labels in dataset.take(1):
        for i in range(25):  # show first 25 images
            ax = plt.subplot(5, 5, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))
            label = int(labels[i].numpy())
            if class_names:
                plt.title(class_names[label])
            else:
                plt.title(str(label))
            plt.axis("off")
    plt.suptitle(title, fontsize=14)
    plt.show()

# Retrieve class names from training directory
class_names = train_ds.class_names
print("Detected classes:", class_names)

# Show sample images from each dataset
show_sample_images(train_ds, "Sample Training Images", class_names)
show_sample_images(val_ds, "Sample Validation Images", class_names)
show_sample_images(test_ds, "Sample Test Images", class_names)
```

```python
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.05),
    layers.RandomZoom(0.08),
    layers.RandomTranslation(0.03, 0.03),
], name="data_augmentation")

def preprocess(x, y):
    x = tf.image.convert_image_dtype(x, tf.float32)
    return x, y

train_ds = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y),
                        num_parallel_calls=AUTOTUNE)
train_ds = train_ds.map(preprocess, num_parallel_calls=AUTOTUNE)
val_ds = val_ds.map(preprocess, num_parallel_calls=AUTOTUNE)
test_ds = test_ds.map(preprocess, num_parallel_calls=AUTOTUNE)

train_ds = train_ds.cache().prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(AUTOTUNE)
test_ds = test_ds.cache().prefetch(AUTOTUNE)
```

```python
# Compute class weights
def compute_class_counts(directory):
    classes = sorted([d for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))])
    counts = {}
    for c in classes:
        counts[c] = len([f for f in os.listdir(os.path.join(directory, c)) if os.path.isfile(os.path.join(directory, c, f))])
    return classes, counts

_, train_counts = compute_class_counts(TRAIN_DIR)
print("Train class counts:", train_counts)

total_train = sum(train_counts.values())
class_weight = {}
for i, cls in enumerate(sorted(train_counts.keys())):
    class_weight[i] = (1 / train_counts[cls]) * (total_train / 2.0)
print("Class weights:", class_weight)
```

#### STAGE-3: MODEL BUILDING

```python
input_shape = IMG_SIZE + (3,)

base_model = efficientnet_v2.EfficientNetV2B0(
    include_top=False,
    weights='imagenet',
    input_shape=input_shape
)
base_model.trainable = False
```

```python
inputs = layers.Input(shape=input_shape)
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(1, activation='sigmoid')(x)

model = models.Model(inputs, outputs, name="EfficientNetV2B0_BreastUS")
```

```python
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=[tf.keras.metrics.BinaryAccuracy(name='accuracy'),
             tf.keras.metrics.Precision(name='precision'),
             tf.keras.metrics.Recall(name='recall'),
             tf.keras.metrics.AUC(name='auc')]
)
```

```python
model.summary()
```

#### STAGE-4: MODEL TRAINING

```python
checkpoint_cb = callbacks.ModelCheckpoint("best_efficientnetv2_b0.h5", save_best_only=True, monitor='val_auc', mode='max')
earlystop_cb = callbacks.EarlyStopping(monitor='val_auc', patience=6, restore_best_weights=True, mode='max')
reduce_lr_cb = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7, verbose=1)
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=[checkpoint_cb, earlystop_cb, reduce_lr_cb],
    class_weight=class_weight
)
```

```python
base_model.trainable = True
fine_tune_at = len(base_model.layers) - 30
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5),
    loss='binary_crossentropy',
    metrics=['accuracy', 'precision', 'recall', 'auc']
)
```

```python
fine_history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=[checkpoint_cb, earlystop_cb, reduce_lr_cb],
    class_weight=class_weight
)
```

```python
def merge_hist(h1, h2):
    merged = {}
    for k in h1.history.keys():
        merged[k] = h1.history[k] + h2.history.get(k, [])
    return merged
merged_history = merge_hist(history, fine_history)
```

```python
def plot_metrics(merged_hist):
    plt.figure(figsize=(14, 5))

    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(merged_hist['accuracy'], label='Train Accuracy')
    plt.plot(merged_hist['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(merged_hist['loss'], label='Train Loss')
    plt.plot(merged_hist['val_loss'], label='Validation Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()
plot_metrics(merged_history)
```

#### STAGE-5: MODEL TRAINING

```python
y_true, y_probs = [], []
for x_batch, y_batch in test_ds:
    preds = model.predict(x_batch, verbose=0)
    y_probs.extend(preds.ravel().tolist())
    y_true.extend(y_batch.numpy().ravel().tolist())

y_true = np.array(y_true)
y_probs = np.array(y_probs)
y_pred = (y_probs >= 0.5).astype(int)
```

```python
acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, zero_division=0)
rec = recall_score(y_true, y_pred, zero_division=0)
f1 = f1_score(y_true, y_pred, zero_division=0)
auc = roc_auc_score(y_true, y_probs)
```

```python
print(f"Test Accuracy : {acc:.4f}")
print(f"Test Precision: {prec:.4f}")
print(f"Test Recall   : {rec:.4f}")
print(f"Test F1-score : {f1:.4f}")
print(f"Test ROC-AUC  : {auc:.4f}")
```

```python
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Benign', 'Malignant'],
            yticklabels=['Benign', 'Malignant'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
fpr, tpr, _ = roc_curve(y_true, y_probs)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
print("Classification Report:")
print(classification_report(y_true, y_pred, digits=4, target_names=['Benign','Malignant']))
```
